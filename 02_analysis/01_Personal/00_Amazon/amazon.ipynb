```python
# Assuming df is your DataFrame and 'Total Owed' is the column you want to convert
# Remove dollar sign and commas (if any), then convert to float
df['Total Owed'] = df['Total Owed'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Assuming df is your DataFrame
filtered_df = df[df['Order Status'] == 'Closed']
filtered_df.head()
```

```python
# Assuming df is your DataFrame
filtered_df['Order Date'] = pd.to_datetime(filtered_df['Order Date'])
```

```python
filtered_df['yearmonth'] = filtered_df['Order Date'].dt.to_period('M')
```

```python
filtered_df['yearmonth'] = filtered_df['yearmonth'].astype(str)
df.head()
```

```python
# Assuming df is your DataFrame
grouped_df = filtered_df.groupby(['yearmonth', 'Shipping Address'])['Total Owed'].sum().reset_index()
grouped_df
```

```python
grouped_df_add = filtered_df.groupby(['Shipping Address'])['Total Owed'].sum().reset_index()
grouped_df_add
```

```python
df_groupby_month = filtered_df.groupby(['yearmonth'])['Total Owed'].sum().reset_index()
df_groupby_month
```

```python
filtered_df['year'] = filtered_df['Order Date'].dt.to_period('Y')
```

```python
filtered_df.head()
```

```python
df_group_year = filtered_df.groupby(['year'])['Total Owed'].sum().reset_index()
```

```python
df_group_year
```

```python
import matplotlib.pyplot as plt

df_groupby_month = filtered_df.groupby(['yearmonth'])['Total Owed'].sum().reset_index()

plt.figure(figsize=(20,10))

plt.plot(df_groupby_month['yearmonth'], df_groupby_month['Total Owed'], marker='o')
plt.title('Total Owed Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Owed')
plt.grid(True)
plt.show()
```

```python
df_group_year['year'] = df_group_year['year'].astype(str)
plt.figure(figsize=(20,10))

plt.plot(df_group_year['year'], df_group_year['Total Owed'], marker='o')
plt.title('Total Owed Over Time')
plt.xlabel('Yearh')
plt.ylabel('Total Owed')
plt.grid(True)
plt.show()
```
